# import libraries

```python
import requests, time, random
from bs4 import BeautifulSoup
from selenium import webdriver
import csv
```

# get the page using url which want to scrap 

```python
res = requests.get('https://int.soccerway.com/');
soup = BeautifulSoup(res.text, 'lxml')
```

```python
content = soup.select('.content')
```

```python
print(content)
```

```python
group_heads = soup.find_all('tr', class_ = 'group-head');
```

# scrap the match list with a link which could give the match details

# after that scrap all the match details for that i make a click in wide button to show details about matches

# Need to connect to your internet connection because every time a new chrome page will open

```python
with open('soccerway.csv', 'w', newline = '') as f:
    thewriter = csv.writer(f)
    
for tr in group_heads:
    match_vs = tr.find('span')
    link = tr.find('a')['href']
    print('match vs: ', match_vs.text)
#     print('Link =   ',link)
    
    base_url = 'https://int.soccerway.com'
    new_url = base_url + link
    
#     cu is define for child url
    
    browser = webdriver.Chrome('C:/Users/anime/Desktop/driver/chromedriver_win32/chromedriver.exe')
    browser.delete_all_cookies()
    browser.get(new_url)
    try:
        browser.find_element_by_id('page_competition_1_block_competition_tables_8_1_3').click()
    except:
        print('not get wide button')
        continue
        
    time.sleep(5)
    src = browser.page_source
    cu_soup = BeautifulSoup(src, 'lxml')
    

    
    content = cu_soup.select('.content')

    matches_sum = cu_soup.find_all('tr', class_ = 'no-date-repetition');
    
    table = soup.select('.block_competition_tables')
    
    for tr in matches_sum:
        day = tr.find('td', class_= 'day')
        date= tr.find('td', class_= 'date')
    
        team_a_link = tr.find('td', class_= 'team-a')
        if(team_a_link):
            team_a = team_a_link.find('a')

        score = tr.find('td', class_ = 'score-time')
        team_b_link = tr.find('td', class_= 'team-b')
        if(team_b_link):
            team_b = team_b_link.find('a')

        if(team_b and team_a and score and date and day):
            print(day.text, date.text, team_a.text, score.text, team_b.text)
            with open('soccerway.csv', 'a', newline = '') as f:
                thewriter = csv.writer(f)
                thewriter.writerow([day.text.replace('/n', '').strip(), date.text.replace('/n', '').strip(), team_a.text.replace('/n', '').strip(), score.text.replace('/n', '').strip(), team_b.text.replace('/n', '').strip()])
#                 thewriter.writerow([day.text.strip('\t\n\r'), date.text.strip('\t\n\r'), team_a.text.strip('\t\n\r'), score.text.strip('\t\n\r'), team_b.text.strip('\t\n\r')])
            
            
    team_ranks = cu_soup.find_all('tr', class_='team_rank')
    
    cnt = 0;
    

    for tr in team_ranks:
        cnt = cnt + 1;
        rank = tr.find('td', class_= 'rank')

        large_link = tr.find('td', class_= 'large-link')
        link = large_link.find('a')['href']
        link = base_url + link
        name = large_link.find('a').text
        
    
        total_mp = tr.find('td', class_='mp')
        total_won = tr.find('td', class_='total_won')
        total_drawn = tr.find('td', class_ = 'total_drawn')
        total_lost = tr.find('td', class_= 'total_lost')
        total_gf = tr.find('td', class_ = 'total_gf')
        total_ga = tr.find('td', class_ = 'total_ga')

        home_total = tr.find('td', class_ = 'home_total')
        home_won = tr.find('td', class_ = 'home_won')
        home_drawn = tr.find('td', class_ = 'home_drawn')
        home_lost = tr.find('td', class_ = 'home_lost')
        home_gf = tr.find('td', class_ = 'home_gf')
        home_ga = tr.find('td', class_ = 'home_ga')

        away_total = tr.find('td', class_ = 'away_total')
        away_won = tr.find('td', class_ = 'away_won')
        away_drawn = tr.find('td', class_ = 'away_drawn')
        away_lost = tr.find('td', class_ = 'away_lost')
        away_gf = tr.find('td', class_ = 'away_gf')
        away_ga = tr.find('td', class_ = 'away_ga')

        gd = tr.find('td', class_='gd')
        points = tr.find('td', class_= 'points')
        
        try:
            with open('soccerway.csv', 'a', newline = '') as f:
                thewriter = csv.writer(f)
                thewriter.writerow([rank.text, link, name, total_mp.text, total_won.text, total_drawn.text, total_lost.text, total_gf.text, total_ga.text, home_total.text, away_total.text, home_won.text, home_drawn.text, home_lost.text, home_gf.text, home_ga.text, away_total.text, away_won.text, away_drawn.text, away_lost.text, away_gf.text, away_ga.text ,gd.text, points.text])
            print(rank.text, link, name, total_mp.text, total_won.text, total_drawn.text, total_lost.text, total_gf.text, total_ga.text, home_total.text, away_total.text, home_won.text, home_drawn.text, home_lost.text, home_gf.text, home_ga.text, away_total.text, away_won.text, away_drawn.text, away_lost.text, away_gf.text, away_ga.text ,gd.text, points.text)
        except:
            print('table row end')
```
